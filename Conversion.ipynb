```python
import astropy.constants as ac
import astropy.units as au
import numpy as np
import pandas as pd
```

```python
def p1_p2(parameter1,type1='eV',type2='nm'):
    """
    Convert the energy, wavelength, and frequency between each other.

    Parameters
    ----------
    parameter1: float
        the number want to be converted
    type1: str
        the units of energy, wavelength, and frequency, 'J'/'eV'/'erg'/'Hz'/'nm'
    type2: str
        the units of energy, wavelength, and frequency, 'J'/'eV'/'erg'/'Hz'/'nm'

    Returns
    -------
    parameter2: float
        the number after converted, with unit type2
        
    """
    const=(ac.h*ac.c).value
    if type1=='J':
        if type2=='eV':
            parameter2=(parameter1*au.J).to(au.eV)
        if type2=='erg':
            parameter2=(parameter1*au.J).to(au.erg)
        if type2=='Hz':
            parameter2=(parameter1/ac.h.value)*au.Hz
        if type2=='nm':
            parameter2=((const/parameter1)*au.m).to(au.nm)
    if type1=='eV':
        if type2=='J':
            parameter2=(parameter1*au.eV).to(au.J)
        if type2=='erg':
            parameter2=(parameter1*au.eV).to(au.erg)
        if type2=='Hz':
            parameter1=((parameter1*au.eV).to(au.J)).value
            parameter2=(parameter1/ac.h.value)*au.Hz
        if type2=='nm':
            parameter1=((parameter1*au.eV).to(au.J)).value
            parameter2=((const/parameter1)*au.m).to(au.nm)
    if type1=='erg':
        if type2=='J':
            parameter2 = (parameter1*au.erg).to(au.J)
        if type2=='eV':
            parameter2 = (parameter1*au.erg).to(au.eV)
        if type2=='Hz':
            parameter1 = ((parameter1*au.erg).to(au.J)).value
            parameter2=(parameter1/ac.h.value)*au.Hz
        if type2=='nm':
            parameter1 = ((parameter1*au.erg).to(au.J)).value
            parameter2=((const/parameter1)*au.m).to(au.nm)
    if type1=='Hz':
        if type2=='J':
            parameter2=parameter1*ac.h.value*au.J
        if type2=='eV':
            parameter2=(parameter1*ac.h.value*au.J).to(au.eV)
        if type2=='erg':
            parameter2=(parameter1*ac.h.value*au.J).to(au.erg)
        if type2=='nm':
            parameter2=((ac.c.value/parameter1)*au.m).to(au.nm)
    if type1=='nm':
        if type2=='J':
            parameter2=(const/(parameter1*au.nm.to(au.m)))*au.J
        if type2=='eV':
            parameter2=((const/(parameter1*au.nm.to(au.m)))*au.J).to(au.eV)
        if type2=='erg':
            parameter2=((const/(parameter1*au.nm.to(au.m)))*au.J).to(au.erg)   
    return parameter2
```
